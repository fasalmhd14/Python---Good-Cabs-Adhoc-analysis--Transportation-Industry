```python
df_trips.head()
```

```python
# df_trip dataframegroupby object by city grouping 
df_trip_cg = df_trips.groupby('city_id')
df_trip_cg
```

```python
#total trips
df_total_trips = df_trip_cg['trip_id'].count()
df_total_trips
```

```python
#average fare per km
```

```python
#average fare per km 
df_avg_far_km = df_trip_cg['fare_amount'].sum() / df_trip_cg['distance_travelled(km)'].sum()
df_avg_far_km = df_avg_far_km.round(2)
df_avg_far_km
```

```python
#average fare per trip 
df_avg_far_trip = df_trip_cg['fare_amount'].sum() / df_trip_cg['trip_id'].count()
df_avg_far_trip = df_avg_far_trip.round(2)
df_avg_far_trip
```

```python
#data validation for the average fare per km and average fare per trip by checking for city == AP01
df_trips[df_trips.city_id == 'AP01'].describe()
```

```python
# the percentage contribution of each city’s trips to the overall trips.
df_per_con = df_trip_cg['trip_id'].count() / df_trips['trip_id'].count()
df_per_con = df_per_con.round(2) * 100
df_per_con
```

```python
df_avg_far_km = df_avg_far_km.reset_index()
df_avg_far_km.rename(columns = {0 : 'Avg Fare Per Km'}, inplace = True)
```

```python
df_avg_far_trip = df_avg_far_trip.reset_index()
df_avg_far_trip.rename(columns = {0 : 'Avg Fare Per Trip'}, inplace = True)
```

```python
df_per_con = df_per_con.reset_index() 
df_per_con.rename(columns = {'trip_id' : 'Perc. Cont'}, inplace = True)
```

```python
df_report_1 = pd.concat([df_total_trips,df_avg_far_km, df_avg_far_trip,df_per_con], axis =1)
df_report_1
```

```python
df_date_trips = df_date_trips.groupby(['start_of_month','city_id'])['trip_id'].count().reset_index()
df_date_trips.rename(columns = {"start_of_month" : "month", "trip_id" : "Actual"}, inplace= True)
```

```python
df_date_trips.head()
```

```python
df_report_2 = pd.merge(df_date_trips, df_target_trips, on = ['month','city_id'])
df_report_2.rename(columns = {"total_target_trips" : "Target"}, inplace= True)
```

```python
def status(x,y):
    if x >= y:
        return "Target Acheived"
    elif x < y:
        return "Below Target"
    
```

```python
df_report_2['Status'] = df_report_2.apply(lambda x: status(x['Actual'], x['Target']), axis =1)
df_report_2.head()
```

```python
df_report_2['Perc. Diff.'] = df_report_2.apply(lambda x: ((x['Actual'] - x['Target'])/x['Target']) * 100, axis =1).round(2)
df_report_2.head()
```

```python
df_pass_city_ag = df_passenger_summary.groupby(['city_id'])['new_passengers'].sum().reset_index()
df_pass_city_ag
```

```python
df_pass_city_ag['rankings'] = df_pass_city_ag.rank(ascending = False, numeric_only = True)
df_pass_city_ag
```

```python
def category(x):
    if x == 1.0 or x == 2.0 or x == 3.0:
        return "Top 3"
    elif x == 8.0 or x == 9.0 or x == 10.0:
        return "Bottom 3"
    else:
        return "other"
```

```python
df_pass_city_ag['category'] = df_pass_city_ag['rankings'].apply(lambda x : category(x))
df_pass_city_ag
```

```python
df_report_4 = df_pass_city_ag[
    (df_pass_city_ag['category'] == "Top 3") | (df_pass_city_ag['category'] == "Bottom 3")
]
df_report_4
```

```python
df_report_4 = pd.merge(df_city, df_report_4, on = "city_id")
df_report_4 = df_report_4.iloc[:,[1,2,4]]
df_report_4.sort_values('new_passengers')
```
